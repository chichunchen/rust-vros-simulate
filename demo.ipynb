```python
def run(level_one, level_two):
    filename = "elephant_96_20_{}_{}.txt".format(level_one, level_two)
    with open(filename, "r") as f:
        index = 1
        wifi_summary = [0, 0, 0, 0]
        soc_summary = [0, 0, 0, 0]
        for line in f:
            line_arr = line.split()
            wifi = float(line_arr[0])
            soc = float(line_arr[1])
            if index % 3 == 1:
                wifi_summary[0] += wifi
                soc_summary[0] += soc
            elif index % 3 == 2:
                wifi_summary[1] += wifi
                soc_summary[1] += soc
            elif index % 3 == 0:
                wifi_summary[2] += wifi
                soc_summary[2] += soc
            index += 1

        size = index / 3
        wifi_summary = list(map(lambda x: x / size, wifi_summary))
        soc_summary = list(map(lambda x: x / size, soc_summary))
        wifi_summary[-1] = 339.1666451915568
        soc_summary[-1] = 421.98720341138164
        
        return wifi_summary, soc_summary
        
wifi_summary, soc_summary = run(1800, 2400)
print(wifi_summary)
print(soc_summary)
```

```python
import matplotlib.pyplot as plt; plt.rcdefaults()
import numpy as np
import matplotlib.pyplot as plt

def draw_histogram(level_one, level_two, wifi, soc):
    objects = ('L1-L2-hier', 'L1-hier', 'pred-L1-L2-miss', 'baseline')
    y_pos = np.arange(len(objects))

    plt.bar(y_pos, wifi, align='center', alpha=0.5)
    plt.xticks(y_pos, objects)
    plt.ylabel('Power Consumption')
    filename = "elephant_96_20_{}_{}.txt".format(level_one, level_two)
    plt.title(filename)

    plt.show()

    objects = ('L1-L2-hier', 'L1-hier', 'pred-L1-L2-miss', 'baseline')
    y_pos = np.arange(len(objects))

    plt.bar(y_pos, soc, align='center', alpha=0.5)
    plt.xticks(y_pos, objects)
    plt.ylabel('Power Consumption')
    plt.title(filename)

    plt.show()
    
draw_histogram(1800, 2400, wifi_summary, soc_summary)
```

```python
wifi_summary, soc_summary = run(1600, 2000)
print(wifi_summary)
print(soc_summary)

draw_histogram(1600, 2000, wifi_summary, soc_summary)
```
